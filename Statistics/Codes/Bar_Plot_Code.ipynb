```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### dataset for barplot

```python
m = 30 #rows
n = 6 #columns

# data generation
data = np.zeros((m,n))

for i in range (n):
    data[:,i] = 30*np.random.randn(m) * (2*i/(n-1)-1)**2 + (i+1)**2
    
```

```python
pd.DataFrame(data)
```

### Plotting  bars

```python
fig,ax = plt.subplots(1,3,figsize=(8,2))

## Only bars
ax[0].bar(range(n),np.mean(data,axis=0))
ax[0].set_title('Bar Plot')

## Only error bars
ax[1].errorbar(range(n),np.mean(data,axis=0),np.std(data,axis=0,ddof=1),marker='s',linestyle='')
ax[1].set_title('Error Plot')

## Ploting both together
ax[2].bar(range(n),np.mean(data,axis=0))
ax[2].errorbar(range(n),np.mean(data,axis=0),np.std(data,axis=0,ddof=1),marker='s',linestyle='',color='black')
ax[2].set_title('Error Bar Plot')
```

If we want to __make group__ with our bars and __keep the groups seperate__ with each other we can do that by making manual x-axis
values


```python
x_crossings = [1,2,4,5,6,9]

plt.bar(x_crossings,np.mean(data,axis=0))
plt.errorbar(x_crossings,np.mean(data,axis=0),np.std(data,axis=0,ddof=1),marker='o',linestyle='',color='black')
plt.title('Error Bar Plot')
plt.show()
```

### More gorgeous visualization of grouped bar plot

```python
m=[[2,5,4,3],[1,1,8,6]]

fig, ax = plt.subplots(2,2,figsize=(8,8))
ax[0,0].imshow(m)


df = pd.DataFrame(m,columns=['prop 0','prop 1','prop 2', 'prop 3'])
df.plot(ax=ax[1,0],kind='bar')
ax[1,0].set_title('Grouping by Rows')


ax[0,1].imshow(np.array(m).T)
df.T.plot(ax=ax[1,1], kind='bar')
ax[1,1].set_title('Grouping by Columns')
```

This code is creating a 2x2 grid of subplots using Matplotlib and then populating these subplots with images and bar plots based on a given matrix `m` and its transposition. It also uses Pandas to create a DataFrame for plotting. Let's break down the code step by step:

1. `m = [[2,5,4,3],[1,1,8,6]]`: This line initializes a 2x4 matrix `m` with two rows and four columns.

2. `fig, ax = plt.subplots(2, 2, figsize=(8, 8))`: This line creates a 2x2 grid of subplots (four subplots in total) using Matplotlib. It assigns the figure object to `fig` and a 2x2 array of Axes objects to `ax`. The `figsize` parameter sets the size of the entire figure to 8x8 inches.

3. `ax[0,0].imshow(m)`: This line displays the matrix `m` as an image in the top-left subplot (ax[0,0]). It essentially treats the values in `m` as pixel intensities, creating a visual representation of the matrix.

4. `df = pd.DataFrame(m, columns=['prop 0','prop 1','prop 2', 'prop 3'])`: Here, a Pandas DataFrame is created from the matrix `m`, with column names 'prop 0', 'prop 1', 'prop 2', and 'prop 3'. This DataFrame will be used for creating bar plots.

5. `df.plot(ax=ax[1,0], kind='bar')`: This line creates a bar plot in the bottom-left subplot (ax[1,0]) using the DataFrame `df`. The `kind='bar'` parameter specifies that you want a bar plot. This will show two bars, one for each row in the DataFrame, and each bar will have four segments corresponding to the columns.

6. `ax[1,0].set_title('Grouping by Rows')`: Sets the title of the bottom-left subplot to 'Grouping by Rows'.

7. `ax[0,1].imshow(np.array(m).T)`: This line displays the transpose of the matrix `m` as an image in the top-right subplot (ax[0,1]). `np.array(m).T` transposes the original matrix, switching rows and columns.

8. `df.T.plot(ax=ax[1,1], kind='bar')`: This line creates a bar plot in the bottom-right subplot (ax[1,1]) using the transpose of the DataFrame `df` (i.e., `df.T`). This will show four bars, one for each column in the original DataFrame, and each bar will have two segments corresponding to the rows.

9. `ax[1,1].set_title('Grouping by Columns')`: Sets the title of the bottom-right subplot to 'Grouping by Columns'.

In summary, this code creates a 2x2 grid of subplots. The top-left subplot displays an image of the original matrix `m`, the bottom-left subplot shows a bar plot of the data in `m` grouped by rows, the top-right subplot displays an image of the transpose of `m`, and the bottom-right subplot shows a bar plot of the data in `m` grouped by columns. It's a visualization of the same data in different ways.
