# Measures of Central Tendancy

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

## MEAN

```python
# Creating 3 dataset

N = 10001
nbins = 30

d1 = np.random.randn(N)-1
d2 = 3*np.random.randn(N)
d3 = np.random.randn(N)+1
```

```python
d1
```

```python
d2
```

```python
d3
```

```python
# histogram 1 data
y1,x1 = np.histogram(d1,nbins)
x1 = (x1[:1]+x1[:-1])/2
```

```python
# histogram 2 data
y2,x2 = np.histogram(d2,nbins)
x2 = (x2[:1]+x2[:-1])/2
```

```python
# histogram 3 data
y3,x3 = np.histogram(d3,nbins)
x3 = (x3[:1]+x3[:-1])/2
```

```python
plt.plot(x1,y1,'b')
plt.plot(x2,y2,'r')
plt.plot(x3,y3,'k')

plt.xlabel('Data values')
plt.ylabel('Data counts')
plt.show()
```

```python
# calculating and plotting the means of the 3 plot

mean_d1 = np.mean(x1)
mean_d2 = np.mean(x2)
mean_d3 = np.mean(x3)
```

```python
plt.plot(x1,y1,'b',x2,y2,'r',x3,y3,'k')
plt.plot([mean_d1,mean_d1],[0,max(y1)],'b--')
plt.plot([mean_d2,mean_d2],[0,max(y2)],'r--')
plt.plot([mean_d3,mean_d3],[0,max(y3)],'k--')

plt.xlabel('Data values')
plt.ylabel('Data counts')
plt.show()
```

### A fail case where mean is failing to show the actual central tendancy of the data

## MEDIAN

```python
# Creating a log normal distribution

shift = 0
stretch = 0.7
n = 10001
nbins = 50

#data
data = stretch*np.random.randn(n) + shift
data = np.exp(data)

```

```python
# histogram data
y,x = np.histogram(data,nbins)
x= (x[1:]+x[:-1])/2
```

```python
mean_data = np.mean(data)
median_data = np.median(data)
```

```python
# plotting data

fig,ax = plt.subplots(2,1,figsize=(4,6))
ax[0].plot(data,'.',color=[.5,.5,.5],label='Data')
ax[0].plot([1,n],[mean_data,mean_data],'r--',label='Mean')
ax[0].plot([1,n],[median_data,median_data],'b--',label='Median')

ax[0].legend()

ax[1].plot(x,y)
ax[1].plot([mean_data,mean_data],[0,max(y)],'r--')
ax[1].plot([median_data,median_data],[0,max(y)],'b--')
ax[1].set_title('Log Normal Data Histogram')
plt.show()
```

## MODE
