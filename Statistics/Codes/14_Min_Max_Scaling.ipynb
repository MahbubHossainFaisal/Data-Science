Min-max scaling is a common technique used in statistics and machine learning to transform numerical data in a way that it falls within a specific range. This range is typically between 0 and 1, but it can be adjusted to any desired minimum and maximum values. Min-max scaling is also known as feature scaling or normalization.

Here's how min-max scaling works:

1. **Select a Range**: Determine the range within which you want to scale your data. The most common choice is to scale it between 0 and 1, but you can choose any minimum and maximum values that make sense for your specific application.

2. **Find the Minimum and Maximum Values**: Calculate the minimum and maximum values of the feature you want to scale within your dataset. Let's call the minimum value Min and the maximum value Max.

3. **Scale the Data**: For each data point in the feature you're scaling (let's call it X), apply the following formula to obtain the scaled value (X_scaled):

   ```
   X_scaled = (X - Min) / (Max - Min)
   ```

   This formula subtracts the minimum value from the data point and then divides it by the range (the difference between the maximum and minimum values). The result is that the scaled data will now fall within the specified range.

4. **Repeat for All Data Points**: Apply the scaling formula to every data point in the feature.

Min-max scaling is useful for several reasons:

- **Normalization**: It ensures that all data points are on the same scale, making it easier to compare and analyze them. This is particularly important when working with machine learning algorithms that are sensitive to the scale of input features, such as gradient descent.

- **Preservation of Relationships**: Min-max scaling preserves the relative relationships between data points. If a data point A is originally greater than a data point B, it will still be greater after scaling, as long as A and B have the same relative positions within the feature.

- **Interpretability**: Scaling data to a common range, like 0 to 1, can make it more interpretable because all values are on the same scale.

However, it's worth noting that min-max scaling may not be suitable for all types of data, especially if your dataset has outliers. In such cases, outliers can disproportionately affect the scaling, causing other data points to be compressed into a very narrow range. In these situations, other scaling techniques like Z-score normalization (standardization) may be more appropriate.

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
# datset

N = 42
data = np.log(np.random.rand(N))*234+294

mindata = min(data)
maxdata = max(data)
````

### Min_Max Scaling

````python
mns = (data-mindata)/(maxdata-mindata)
mns
````

### Plotting and comparing the normal distribution and min_max scaling

````python
fig,ax = plt.subplots(1,2,figsize=(8,4))

ax[0].plot(1+np.random.randn(N)/20,data,'ks')
ax[0].set_xlim([0,2])
ax[0].set_xticks([])
ax[0].set_ylabel('Original data scale')
ax[0].set_title('Original data')


ax[1].plot(1+np.random.randn(N)/20,mns,'ks')
ax[1].set_xlim([0,2])
ax[1].set_xticks([])
ax[1].set_ylabel('Unity Normal Data Scale')
ax[1].set_title('Scaled data')
````

### Scaling doesn't effect the relative values

````python
plt.plot(data,mns,'ks')
plt.xlabel('Original Data')
plt.ylabel('Scaled Data')
plt.show()
````
