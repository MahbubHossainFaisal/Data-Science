## Pie Chart

__Nominal__, __Ordinal__ and __Discrete__ Data can be shown in pie charts

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Preparing data for pie charts

```python
nbins = 5
totalN = 100

data = np.ceil(np.logspace(np.log10(1/2),np.log10(nbins-0.1),totalN))
```

```python
data
```

```python
# Preparing data for the pie chart

unq = np.unique(data)
pieData = np.zeros(len(unq))

for i in range(len(unq)):
    pieData[i] = sum(data == unq[i])
```

```python
pieData
```

```python
plt.pie(pieData, labels= 100*pieData/sum(pieData))
plt.show()
```

### Exploding Pies

```python
plt.pie(pieData, labels= 100*pieData/sum(pieData), explode=[0,0.1,0.2,0.1,0])
plt.show()
```

## For Continuous Data

```python
# Generate log normal distribution

data = np.exp(np.random.randn(1000)/10)
```

```python
data
```

```python
# Generate bins using histogram

hist = np.histogram(data,bins=6)

hist
```

```python
# Show that as a pie chart

plt.pie(hist[0],labels=hist[0])
plt.show()
```
