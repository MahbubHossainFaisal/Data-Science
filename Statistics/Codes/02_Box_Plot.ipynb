```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
m = 30 #rows
n = 6 #columns

# data generation
data = np.zeros((m,n))

for i in range (n):
    data[:,i] = 30*np.random.randn(m) * (2*i/(n-1)-1)**2 + (i+1)**2
    
```

### Basic box plot

```python


plt.boxplot(data)
plt.show()
```

### box plot for the transposed data

```python
plt.boxplot(data.T)
plt.show()
```

### Boxplot with Seaborn

```python
# Vertical

sns.boxplot(data=data,orient='v')
plt.show()
```

```python
# horizontal

sns.boxplot(data=data,orient='h')
plt.show()
```

### Boxplot using Pandas DataFrame

```python
df = pd.DataFrame(data, columns=['zero','one','two','three','four','five'])
df
```

```python
sns.boxplot(data = df,orient='v')
plt.show()
```

```python
sns.boxplot(data = df,orient='h')
plt.show()
```

### Taking a normally distributed data and a uniformly distributed data and plotting their box plot together

```python
mean = 0  # Mean of the distribution
std_dev = 1  # Standard deviation of the distribution
num_samples = 100  # Number of data points to generate

# normally distributed data
data1 = np.random.normal(mean, std_dev, num_samples)
```

```python
low = 0  # Lower bound of the distribution
high = 1  # Upper bound of the distribution
num_samples = 100  # Number of data points to generate

# Uniformly distributed Data
data2 = np.random.uniform(low, high, num_samples)
```

```python
df = pd.DataFrame({'Normal': data1.flatten(), 'Uniform': data2.flatten()})
df
```

```python
sns.boxplot(data=df,orient='v')
plt.show()
```



| Aspect                         | Normally Distributed Data                 | Uniformly Distributed Data                 |
|--------------------------------|------------------------------------------|--------------------------------------------|
| **Probability Density**        | Bell-shaped curve (Gaussian distribution) | Flat and constant probability density     |
| **Shape of Distribution**      | Symmetrical around the mean              | Symmetrical across the entire range        |
| **Mean (Average)**             | Centered around a specified mean         | Equally distributed across the range       |
| **Standard Deviation**         | Describes the spread of data             | Irrelevant, as data has constant density   |
| **Outliers**                   | Less likely due to central tendency      | Equally likely throughout the range        |
| **Data Range**                 | Extends to infinity in both directions   | Limited to a specific range                |
| **Example Applications**       | Height measurements, IQ scores, errors   | Random coin flips, dice rolls, lottery numbers |
