```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import seaborn as sns
```

## Manual creation violine plot

```python
n = 1000
thresh = 5 # Threshold for cropping the data

data = np.exp(np.random.randn(n))
data[data>thresh] = thresh + np.random.randn(sum(data > thresh))*0.1

plt.hist(data,30)
plt.title('Histogram')
plt.show()

# violine plot
plt.violinplot(data)
plt.title('Violin Plot')
plt.show()

```

We can say that violine plots are made from two histograms which are rotated.

## Plotting violin plot using seaborn

## Comparing two dataset distribution using Violin plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Example data
data_x = [1, 2, 3, 4, 5]
data_y = [5, 4, 3, 2, 1]

# Create a DataFrame for the data
df = pd.DataFrame({'Value': data_x + data_y, 'Dataset': ['X'] * len(data_x) + ['Y'] * len(data_y)})

plt.figure(figsize=(10, 5))  # Adjust the figure size as needed

# Create a combined violin plot for both datasets
sns.violinplot(x='Dataset', y='Value', data=df, palette='muted', split=False)

plt.xlabel("Dataset")
plt.ylabel("Value")
plt.title("Violin Plot of Two Datasets (X and Y)")
plt.show()
```
