````python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
````

````python
# Creating 3 dataset

N = 10001
nbins = 30

d1 = np.random.randn(N)-1
d2 = 3*np.random.randn(N)
d3 = np.random.randn(N)+1
````

````python
# histogram 1 data
y1,x1 = np.histogram(d1,nbins)
x1 = (x1[:1]+x1[:-1])/2
````

````python
# histogram 2 data
y2,x2 = np.histogram(d2,nbins)
x2 = (x2[:1]+x2[:-1])/2
````

````python
# histogram 3 data
y3,x3 = np.histogram(d3,nbins)
x3 = (x3[:1]+x3[:-1])/2
````

````python
plt.plot(x1,y1,'b')
plt.plot(x2,y2,'r')
plt.plot(x3,y3,'k')

plt.xlabel('Data values')
plt.ylabel('Data counts')
plt.show()
````

````python
# calculating and plotting the means of the 3 plot

mean_d1 = np.mean(x1)
mean_d2 = np.mean(x2)
mean_d3 = np.mean(x3)
````

### Standard Deviation

````python
stds = np.zeros(3)
stds
````

````python
stds[0] = np.std(d1,ddof=1)
stds[1] = np.std(d2,ddof=1)
stds[2] = np.std(d3,ddof=1)
````

````python
# plotting Mean
plt.plot(x1,y1,'b',x2,y2,'r',x3,y3,'k')
plt.plot([mean_d1,mean_d1],[0,max(y1)],'b--')
plt.plot([mean_d2,mean_d2],[0,max(y2)],'r--')
plt.plot([mean_d3,mean_d3],[0,max(y3)],'k--')

# plotting standard deviation

plt.plot([mean_d1-stds[0],mean_d1+stds[0]],[.16*max(y1),.16*max(y1)],'b',linewidth=8)
plt.plot([mean_d2-stds[1],mean_d2+stds[1]],[.12*max(y2),.12*max(y2)],'r',linewidth=8)
plt.plot([mean_d3-stds[2],mean_d3+stds[2]],[.2*max(y3),.2*max(y3)],'k',linewidth=8)
plt.xlabel('Data values')
plt.ylabel('Data counts')
plt.show()
````

 ## Measures of Different Variances

````python
variances = np.arange(1,11)
N=3000

varmeasures = np.zeros((4,len(variances)))
varmeasures
````

````python
for i in range(len(variances)):
    # create data and center the mean
    data = np.random.randn(N) * variances[i]
    center_data = data - np.mean(data)
    
    # variance
    varmeasures[0,i] = sum(center_data**2)/(N-1)
    
    # Biased Variance
    varmeasures[1,i] = sum(center_data**2)/N
    
    # Standard Deviation
    varmeasures[2,i] = np.sqrt(sum(center_data**2)/(N-1))
    
    # Mean Absolute Difference (MAD)
    varmeasures[3,i] = sum(abs(center_data))/(N-1)
    
    
    
# plotting the variances

plt.plot(variances,varmeasures.T)
plt.legend(('Var','Biased Var','STD','MAD'))
plt.show()
````

In summary, this code generates random data with varying levels of variance, calculates four different measures of variance (sample variance, biased variance, standard deviation, and mean absolute difference) for each variance value, and then plots these measures against the range of variance values. This can help visualize how these variance measures behave as the underlying data's variance changes.

The variance plot in the provided code describes how four different measures of variability (variance) change as the underlying data's variance is manipulated. Each measure provides a different perspective on how spread out or concentrated the data points are. Here's what each of the four measures represents:

1. **Sample Variance (`Var`)**: This is the most common measure of variance. It quantifies the spread or dispersion of data points around the mean. A higher sample variance indicates that data points are more spread out from the mean.

2. **Biased Variance (`Biased Var`)**: This is similar to the sample variance but uses `N` (the total number of data points) in the denominator instead of `(N - 1)` when calculating the average squared differences from the mean. This can result in a slightly different value and is often used when you have the entire population data.

3. **Standard Deviation (`STD`)**: The standard deviation is the square root of the variance. It provides a measure of how much individual data points deviate from the mean. A higher standard deviation indicates greater variability.

4. **Mean Absolute Difference (`MAD`)**: The Mean Absolute Difference measures the average absolute difference between each data point and the mean. It is another way to quantify the spread of data points, emphasizing the magnitude of differences rather than squared differences.

The plot displays how each of these measures changes as the variance of the generated data points is altered by the `variances` array. By varying the level of variance in the data, you can observe how these different measures of variability respond. For example, you might see that as the variance increases, all four measures also increase, indicating greater spread or dispersion of data points.

This type of plot is useful for understanding how different measures of variability are affected by changes in data characteristics and can be used for statistical analysis and data exploration. It helps in visualizing how data spreads or clusters around its mean and provides insights into the shape and nature of the data distribution.
