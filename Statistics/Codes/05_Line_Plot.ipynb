## Line Plot VS Bar Plot

Bar plots and line plots are two common types of data visualization techniques, and they each have their own strengths and are appropriate for different scenarios. Here's a comparison of when to use bar plots and when to use line plots, along with examples:

**Bar Plots:**

1. **Use Cases**:
   - Bar plots are suitable for displaying and comparing discrete categories or groups of data.
   - They are effective for showing the distribution of categorical or nominal data.
   - Bar plots are often used for visualizing counts, frequencies, or proportions.

2. **Examples**:
   - **Comparison of Sales by Product Categories**: You can use a bar plot to compare the sales of different product categories (e.g., electronics, clothing, food) for a given time period.
   - **Survey Responses**: When analyzing the results of a survey with response categories (e.g., strongly agree, agree, neutral, disagree, strongly disagree), a bar plot can show the distribution of responses.
   - **Market Share**: Bar plots are excellent for illustrating the market share of different companies or products within an industry.

**Line Plots:**

1. **Use Cases**:
   - Line plots are ideal for showing trends, patterns, and changes over continuous or sequential data.
   - They are useful for visualizing time series data or data where the x-axis represents a continuous variable (e.g., time, temperature, distance).
   - Line plots can reveal relationships and patterns in data over a range of values.

2. **Examples**:
   - **Stock Price Over Time**: Line plots are commonly used to display the historical prices of a stock over a period of time, showing how the price changes continuously.
   - **Temperature Variation Throughout the Day**: Line plots can represent how temperature changes throughout the day, with time on the x-axis and temperature on the y-axis.
   - **Population Growth Over Years**: Line plots can illustrate how a population grows or declines over multiple years.

**Key Differences**:

- **Bar plots** use discrete bars to represent categories or groups, and there are gaps between the bars to indicate that the categories are distinct.
- **Line plots** use lines to connect data points along a continuous or sequential x-axis. The lines imply a continuous relationship or progression between the data points.

**Choosing Between Bar Plots and Line Plots**:

- Choose a **bar plot** when you want to compare distinct categories or groups, or when you have categorical data with no inherent order.
- Choose a **line plot** when you want to visualize trends, patterns, or changes over continuous or sequential data, especially when there is a natural progression along the x-axis.

Ultimately, the choice between bar plots and line plots depends on the nature of your data and the story you want to convey through your visualization.

### Logarithmic Scale Plot VS Linear Scale Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creating dataset

n= 1000

data1 = np.exp(np.random.randn(n)/2)
data2 = np.exp(np.random.randn(n)/10)
data3 = np.exp(np.random.randn(n)/2+1)
```

```python
# number of bins
k=20

plt.hist(data1,bins=k)
plt.hist(data2,bins=k)
plt.hist(data3,bins=k)
plt.show()
```

We can't see all the histograms properly. In this case, line plot will be really useful

```python
y1,x1 = np.histogram(data1,bins=k)
xx1 = (x1[0:-1] + x1[1:])/2 # bin centering for creating line charts smoothly
y1 = 100*y1/sum(y1) # converting into probability
```

```python
y2,x2 = np.histogram(data2,bins=k)
xx2 = (x2[0:-1] + x2[1:])/2 # bin centering for creating line charts smoothly
y2 = 100*y2/sum(y2) # converting into probability
```

```python
y3,x3 = np.histogram(data3,bins=k)
xx3 = (x3[0:-1] + x3[1:])/2 # bin centering for creating line charts smoothly
y3 = 100*y3/sum(y3) # converting into probability
```

```python
plt.plot(xx1,y1,'s-',label='data1')
plt.plot(xx2,y2,'o-',label='data2')
plt.plot(xx3,y3,'^-',label='data3')

plt.legend()
plt.xlabel('Value')
plt.ylabel('Probability')
plt.show()
```

Linear vs Logarithmic Scale Graphs

```python
import matplotlib.pyplot as plt
import numpy as np

# Create an array of x-values
x = np.linspace(0.1, 5, 1000)  # Adjust the range and number of points as needed

# Calculate y-values for y = x and y = e^x
y_linear_x = x
y_linear_exp = np.exp(x)
y_log_x = x
y_log_exp = np.exp(x)

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot y = x and y = e^x on the linear y-axis
ax1.plot(x, y_linear_x, label='y = x', color='blue')
ax1.plot(x, y_linear_exp, label='y = e^x', color='green')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend()

# Plot y = x and y = e^x on the logarithmic y-axis
ax2.plot(x, y_log_x, label='y = x', color='blue')
ax2.plot(x, y_log_exp, label='y = e^x', color='green')
ax2.set_xlabel('x')
ax2.set_ylabel('y (log scale)')
ax2.set_yscale('log')
ax2.legend()

# Title for the combined plot
fig.suptitle('Linear and Logarithmic Y-Axis')

# Display the plots
plt.tight_layout()
plt.show()
```
