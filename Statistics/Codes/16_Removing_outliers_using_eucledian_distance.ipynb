```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# dataset

n = 40

d1 = np.exp(-abs(np.random.randn(n)*3))
d2 = np.exp(-abs(np.random.randn(n)*5))
datamean = [np.mean(d1),np.mean(d2)]
```

### Computing data of each point to mean

```python
ds = np.zeros(n)

for i in range(n):
    ds[i] = np.sqrt((d1[i]-datamean[0])**2 + (d2[i]-datamean[1])**2)
    
ds
```

### Converting to z score

```python
zdata = (ds-np.mean(ds)/np.std(ds))
zdata
```

```python
## plotting data

fig, ax = plt.subplots(1,2,figsize=(8,6))

ax[0].plot(d1,d2,'ko',markerfacecolor='k')
ax[0].set_xticks([])
ax[0].set_yticks([])
ax[0].set_xlabel('Variable x')
ax[0].set_ylabel('Variable y')

# ploting the multivariate mean
ax[0].plot(datamean[0],datamean[1],'kp',markerfacecolor='g',markersize=12)


# Plotting the distance
ax[1].plot(zdata,'ko',markerfacecolor=[.7,.5,.3],markersize=12)
ax[1].set_xlabel('Data Index')
ax[1].set_ylabel('Z distance')
```

### Detecting outliers by checking which eucledian distance is greater from mean

```python
# threshold assumption
thresh = -1.2
indx = np.where((zdata) > thresh)[0]

print(indx)
```

```python
# plot

ax[1].plot(indx,zdata[indx],'x',color='r',markersize=20)
ax[0].plot(d1[indx],d2[indx],'x',color='r',markersize=20)

fig
```
