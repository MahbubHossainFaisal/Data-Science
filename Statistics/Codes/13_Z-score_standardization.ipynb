Z-score standardization, also known as the standard score or normal deviate, is a statistical technique used to transform data into a standard normal distribution with a mean (average) of 0 and a standard deviation of 1. This transformation is helpful in various statistical analyses and machine learning tasks because it allows you to compare and work with data on a common scale. Here's how you calculate and use the Z-score:

**Step 1: Calculate the Mean and Standard Deviation**

Before you can compute the Z-score for a dataset or a specific data point, you need to calculate the mean (μ) and the standard deviation (σ) for that dataset. The mean represents the average value of the dataset, while the standard deviation measures the spread or variability of the data.

**Step 2: Calculate the Z-Score**

Once you have the mean and standard deviation, you can calculate the Z-score for a data point (x) using the following formula:

Z = (x - μ) / σ

Where:
- **Z** is the Z-score.
- **x** is the data point you want to standardize.
- **μ** is the mean of the dataset.
- **σ** is the standard deviation of the dataset.

**Step 3: Interpret the Z-Score**

The resulting Z-score tells you how many standard deviations a data point is from the mean. Here's what the Z-score means in different scenarios:

- A Z-score of 0 means the data point is exactly at the mean.
- A positive Z-score indicates that the data point is above the mean.
- A negative Z-score indicates that the data point is below the mean.

The magnitude of the Z-score tells you how far the data point is from the mean in terms of standard deviations. For example:
- A Z-score of 1 means the data point is 1 standard deviation above or below the mean.
- A Z-score of 2 means the data point is 2 standard deviations above or below the mean.

**Step 4: Standardized Data**

Once you have computed the Z-scores for all data points in your dataset, you will have a standardized dataset where the mean is 0 and the standard deviation is 1. This transformation can be very useful in various statistical analyses and modeling tasks, as it simplifies comparisons and helps in identifying outliers.

**Applications of Z-Score Standardization:**

1. **Outlier Detection:** Z-scores are often used to identify outliers in a dataset. Data points with Z-scores significantly above or below a certain threshold may be considered outliers.

2. **Normalization for Machine Learning:** Standardizing data is a common preprocessing step in machine learning. Algorithms that rely on distance measures, like k-means clustering or principal component analysis (PCA), benefit from standardized data because it ensures that all features are on the same scale.

3. **Hypothesis Testing:** Z-scores are used in hypothesis testing, especially in z-tests and t-tests, to determine whether sample means are significantly different from population means.

4. **Quality Control:** In manufacturing and quality control, Z-scores are used to assess the quality and consistency of products.

In summary, Z-score standardization is a valuable technique for transforming data to have a mean of 0 and a standard deviation of 1, making it easier to analyze and compare data across different scales and datasets.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
# dataset

data = np.random.poisson(3,1000)**2
```

```python
# mean
data_mean = np.mean(data)
data_mean
```

```python
# std
data_std = np.std(data,ddof=1)
data_std
```

## Plotting the data

```python
plt.plot(data,'s',markersize=3)
plt.xlabel('Data Index')
plt.ylabel('Data Value')
plt.title(f'Mean: {np.round(data_mean)} | STD: {np.round(data_std)}')
plt.show()
```

## Applying Z score standardization to the dataset

```python
z_data = stats.zscore(data)
```

```python
# commputing mean and std of z score standardized data
z_mean = np.mean(z_data)
z_mean
```

```python
z_std = np.std(z_data)
z_std
```

```python
plt.plot(z_data,'s',markersize=3)
plt.xlabel('Data Index')
plt.ylabel('Data Value')
plt.title(f'Mean: {np.round(z_mean)} | STD: {np.round(z_std)}')
plt.show()
```

## Plotting correlation between the data and z transformed data

```python
plt.plot(data,z_data,'s')
plt.xlabel('Original')
plt.ylabel('Z Transformed')
plt.title(f'Correlation between original and z transformed data')
plt.show()
```
