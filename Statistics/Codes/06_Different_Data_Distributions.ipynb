```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

## Gaussian Distribution (Normal Distribution)

```python
# number of discretizations
n = 1001

x = np.linspace(-4,4,n)

gausdist = stats.norm.pdf(x)

plt.plot(x,gausdist)
plt.title("Gaussian (Normal) Distribution")
plt.show()
```

### Is this a probabibility distribution function?

```python
print(sum(gausdist))
```

The value should be 1. So it is not a probability Distribution function

### Convert it to a probability distribution function

```python
# number of discretizations
n = 1001

x = np.linspace(-4,4,n)

gausdist = stats.norm.pdf(x)

plt.plot(x,gausdist/sum(gausdist))
plt.title("Analytic Gaussian (Normal) Distribution") # Analytic because we have generated it from a formula
plt.show()
```

```python
print(sum(gausdist)/sum(gausdist))
```

Now it is a probability distribution as the sum of the pdf values is 1.

## Emperical Normal Distribution ( Create your own Normal Distribution )

```python
# parameters
strech = 1 #variance
shift = 5 #mean
n = 1000

# create dataset
data = strech*np.random.randn(n) + shift
```

```python
plt.hist(data,25)
plt.title('Emperical Normal Distribution')
plt.show()
```

## Uniformly Distributed Data

```python
# parameters

stretch = 2 #variance
shift = .5 #mean
n = 10000

data = strech*np.random.rand(n) + shift-strech/2
```

```python
fig,ax = plt.subplots(2,1,figsize=(5,5))

ax[0].plot(data,'.',markersize=1)
ax[0].set_title('Uniform data value')

ax[1].hist(data,25)
ax[1].set_title('Uniform data histogram')

plt.show()
```

## Analytic Log-Normal Distribution

```python
# parameters

N = 1001
x = np.linspace(0,10,N)
lognormdist = stats.lognorm.pdf(x,1)
```

```python
plt.plot(x,lognormdist)
plt.title('Analytic Log-Normal Distribution')# Analytic because we have generated it from a formula
plt.show()
```

## Emperical Log-Normal Distribution

```python
# parameters

shift = 5
strech = 0.5
n = 2000 

# dataset
data = np.exp(strech*np.random.randn(n) + shift)
```

```python
fig,ax = plt.subplots(2,1,figsize=(4,6))
ax[0].plot(data,'.')
ax[0].set_title('Log-Normal Data values')

ax[1].hist(data,25)
ax[1].set_title('Log-Normal Data Histogram')

plt.show()
```

## Binomial Distribution

```python
# it is used to show the probability distribution of k heads in N coin tosses and similar scenarios

n = 10 #number of coins
p = 0.5 #probability of getting heads

x = range(n+2)
binodist = stats.binom.pmf(x,n,p)

plt.bar(x,binodist)
plt.title('Binomial Distribution (n=%s,p=%g)'%(n,p))
plt.xlabel('Number of coins')
plt.ylabel('Probability Density')
plt.show()
```

## t Dirstribution

```python
# parameters

x = np.linspace(-4,4,1001)
df = 200 # degree of freedom

t = stats.t.pdf(x,df)
```

```python
plt.plot(x,t)
plt.xlabel('t-Value')
plt.ylabel('P(t | H$_0$)')
plt.title('t(%g) distribution'%df)
plt.show()
```

## F Distribution

```python
# parameters

num_df = 5 #numerator degree of freedom
den_df = 100 #denominator degree of freedom

# data
x = np.linspace(0,10,10001)

fdist = stats.f.pdf(x,num_df,den_df)
```

```python
plt.plot(x,fdist)
plt.title(f'F{num_df},{den_df} Distribution')
plt.xlabel('F Value')
plt.show()
```
