## Computing Derivatives with Differentiation

Equation $y = x^2 + 2x + 2$:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,10000)  # start, finish, n points
```

```python
def func(x_val):
    y_val = x_val**2 + 2*x_val + 2
    
    return y_val
```

```python
y = func(x)
```

```python
fig,ax = plt.subplots()

plt.axvline(x=0,color='green')
plt.axhline(y=0,color='green')
_ = ax.plot(x,y)
```

Let's identify the slope where, say, $x = 2$.

First, let's determine what $y$ is: 

```python
func(2)
```

Let's say this point $P$, which is located at (2, 10):

```python
# showing the P point in the graph
fig,ax = plt.subplots()

plt.axvline(x=0,color='green')
plt.axhline(y=0,color='green')
plt.scatter(2,10) # This is the P point
_ = ax.plot(x,y)
```

The _delta method_ uses the difference between two points to calculate slope. To illustrate this, let's define another point, $Q$ where, say, $x = 4$.

```python
# lets check the y value for x = 4
func(4)
```

Let's say this point $Q$, which is located at (4, 26)

```python
# showing the Q point in the graph
fig,ax = plt.subplots()

plt.axvline(x=0,color='green')
plt.axhline(y=0,color='green')
plt.scatter(2,10,c='red') # This is the P point
plt.scatter(4,26,c='red') # This is the Q point
ax.plot(x,y)
plt.show()
```

To find the slope $m$ between points $P$ and $Q$: 
$$m = \frac{\text{change in }y}{\text{change in }x} = \frac{\Delta y}{\Delta x} = \frac{y_2 - y_1}{x_2 - x_1} = \frac{26-10}{4-2} = \frac{16}{2} = 8$$

```python
# calculating slope
m = (26-10)/(4-2)
m
```

To plot the line that passes through $P$ and $Q$, we can rearrange the equation of a line $y = mx + b$ to solve for $b$: 
$$b = y - mx$$

```python
# calculating b

b = 26-(8*4) # y2=26,x2=4
b
```

```python
# Creating the straight line
line_y = m*x + b
line_y
```

```python
# Ploting the slope line

fig,ax = plt.subplots()

plt.axvline(x=0,color='green')
plt.axhline(y=0,color='green')
plt.scatter(2,10,c='red') # This is the P point
plt.scatter(4,26,c='red') # This is the Q point
plt.plot(x,line_y,c='black')
plt.ylim(-10,150) # making the y axix shorter
ax.plot(x,y)
plt.show()
```

The closer $Q$ becomes to $P$, the closer the slope $m$ comes to being the true tangent of the point $P$. Let's demonstrate this with another point $Q$ at $x = 2.1$.

Previously, our $\Delta x$ between $Q$ and $P$ was equal to 2. Now it is much smaller: $$\Delta x = x_2 - x_1 = 2.1 - 2 = 0.1 $$

```python
# lets check the y value for x = 2.1
func(2.1)
```

The point $Q$, which is now located at (2.1, 10.61)

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, 10)
plt.scatter(2.1, 10.61, c = 'orange', zorder=3)
_ = ax.plot(x,y)
```

```python
# new slop
m = (10.61-10)/(2.1-2)
m
```

```python
# calculating b for the new straightline
b = 10.61-m*2.1
b
```

```python
# straight line (slope)
line_y = m*x + b
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='green')
plt.axhline(y=0, color='green')
plt.scatter(2, 10,c='black')
plt.scatter(2.1, 10.61, c='magenta', zorder=3)
plt.ylim(-5, 150)
plt.plot(x, line_y, c='orange', zorder=3)
_ = ax.plot(x,y)
```

The closer $Q$ becomes to $P$ (i.e., $\Delta x$ approaches 0), the clearer it becomes that the slope $m$ at point $P$ = (2, 10) is equal to 6.

Let's make $\Delta x$ extremely small, 0.000001, to illustrate this:

```python
delta_x = 0.000001
delta_x
```

```python
x1 = 2
y1 = 10
```

Rearranging $\Delta x = x_2 - x_1$, we can calculate $x_2$ for our point $Q$, which is now extremely close to $P$: 
$$x_2 = x_1 + \Delta x$$

```python
x2 = x1 + delta_x
x2
```

$y_2$ for our point $Q$ can be obtained with the usual function $f(x)$: 
$$y_2 = f(x_2)$$

```python
y2 = func(x2)

y2
```

So, our new point Q is located at (2.000001 , 10.000006000001001) point

To find the slope $m$, we continue to use $$m = \frac{\Delta y}{\Delta x} = \frac{y_2 - y_1}{x_2 - x_1}$$

```python
# slope

m = (y2 - y1)/(x2 - x1)
m
```

Using the delta method, we've seen that at point $P$, the slope of the curve is 6.

Using the delta method, we've derived the definition of differentiation from first principles. The derivative of $y$ (denoted $dy$) with respect to $x$ (denoted $dx$) can be represented as: 
$$\frac{dy}{dx} = \lim_{\Delta x \to 0} \frac{\Delta y}{\Delta x}$$

Expanding $\Delta y$ out to $y_2 - y_1$: 
$$\frac{dy}{dx} = \lim_{\Delta x \to 0} \frac{y_2 - y_1}{\Delta x}$$

Finally, replacing $y_1$ with $f(x)$ and replacing $y_2$ with $f(x + \Delta x)$, we obtain a common representation of differentiation:
$$\frac{dy}{dx} = \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x}$$
