$$\lim_{x \to 1} \frac{x^2 - 1}{x - 1}$$

```python
def problem1(x_val):
    y_val = (x_val**2 - 1) / (x_val - 1)
    return y_val
```

## Problem 2

$$\lim_{x \to 0} \frac{\text{sin } x}{x}$$

```python
def func(x_val):
    y_val = np.sin(x_val)/x_val
    
    return y_val
```

```python
func(0.1)
```

```python
func(0.001)
```

```python
func(0)
```

```python
func(-0.1)
```

```python
func(1.1)
```

```python
func(-0.00001)
```

```python
func(-0.99999)
```

```python
y = func(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-10, 10)
plt.ylim(-1, 2)
plt.axvline(x=0, color='purple', linestyle='--')
plt.axhline(y=1, color='purple', linestyle='--')
_ = ax.plot(x,y)
```

## Problem 3

$$ \lim_{x \to \infty} \frac{25}{x} $$

```python
def func(x_val):
    y_val = 25/x_val
    
    return y_val
```

```python
func(0.1)
```

```python
func(0)
```

```python
func(0.01)
```

```python
func(0.001)
```

```python
func(1)
```

```python
func(1.1)
```

```python
func(9)
```

```python
y = func(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-10, 10)
plt.ylim(-300, 300)
_ = ax.plot(x, y)
```

Here We can see a connection between the x and y which is that vertical blue line in (0.0). Now we will take x and y values
seperately so that there is no connections between them

```python
left_x = x[x<0]
right_x = x[x>0]
```

```python
left_y = func(left_x)
right_y = func(right_x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-10, 10)
plt.ylim(-300, 300)
ax.plot(left_x, left_y, c='green')
_ = ax.plot(right_x, right_y, c='blue')
```
