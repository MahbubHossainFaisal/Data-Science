````python
#before converting missing values
Pclass_group['Embarked'].value_counts()
````

````python
# code here
def fillMissingValues(x):
  return  x.fillna(x.mode().iloc[0])

titanic_df['Embarked'] = Pclass_group['Embarked'].transform(fillMissingValues)
````

the transform() method is used to apply the fill_with_mode() function to each group defined by the 'Pclass' column. The transform() method returns a Series with the same length as the original DataFrame, allowing you to update the 'Embarked' column with the modified values within each group.

###`Q-5:` Find batsman in below category-
* Highest score while chasing
* Best Strike rate while chasing and have faced 100+ balls


> Chasing mean team batting in second inning

````python
# code here
chasing_summary = ipl[ipl['innings'] == 2]
chasing_summary.head(5)
````

````python
# Highest score while chasing
batter_group = chasing_summary.groupby('batter')

batter_group['batsman_run'].sum().sort_values(ascending=False).values[0]
````

````python
#Best Strike rate while chasing and have faced 100+ balls\

# finding total run and number of ball played
dic={
    'batsman_run':'sum',
    'ballnumber': 'count'
}
batsman_stats = batter_group.agg(dic)


batsman_stats
````

````python
# calculating strike rate and adding into columns
batsman_stats['Strike rate'] = (batsman_stats['batsman_run'] / batsman_stats['ballnumber']) * 100

batsman_stats
````

````python
type(batsman_stats)
````

````python
batsman_stats[batsman_stats['ballnumber'] > 100]['Strike rate'].sort_values(ascending=False).values[0].round(2)
````

###`Q-6` Most Successful bowler against any batsman. Find that pair of bowler and batsman.
> Most Successful in terms of dissmissal. A bowler who have dissmissed any batsman most no of times. If any two pairs have same no of dissmisal, consider runs conceded by bowler to that batsman. Those who have concede lesser runs is more successful.

````python
# Wicket deliveries
wicket_deliveries = ipl[ipl['isWicketDelivery'] == 1]
wicket_deliveries
````

````python
# bowler that takes the wicket and player who is out

dismissals = wicket_deliveries.groupby(['bowler', 'batter']).size().reset_index(name='dismissals')
dismissals
````

````python
# runs conceded by the bowler against each batsman
runs_conceded = wicket_deliveries.groupby(['bowler', 'batter'])['total_run'].sum().reset_index(name='runs_conceded')
runs_conceded.sort_values('runs_conceded',ascending=False)
````

````python
# Merge dismissals and runs_conceded DataFrames
combined = dismissals.merge(runs_conceded, on=['bowler', 'batter'])
combined
````

````python
# pair with the highest number of dismissals and least runs conceded
most_successful_dismissals = combined.sort_values(['dismissals', 'runs_conceded'], ascending=[False, True]).iloc[0]
most_successful_dismissals
````

````python
# pair with the least runs conceded and at least one dismissal
least_runs_conceded = combined[combined['dismissals'] > 0].sort_values('runs_conceded').iloc[0]
least_runs_conceded
````

###`Q-8:` Make a dataframe for all batting pairs played together.
```
Batsman1 Batsman2 Runs Avg StrikeRate
```

> Just to ease this question you can count wide-balls for strike rate.

````python
# total runs scored by each batting pair
batting_pairs = ipl.groupby(['batter', 'non-striker'])['total_run'].sum().reset_index(name='Runs')

# number of innings played by each batting pair
innings_played = ipl.groupby(['batter', 'non-striker'])['innings'].nunique().reset_index(name='InningsPlayed')

# average runs per innings for each batting pair
batting_pairs['Avg'] = batting_pairs['Runs'] / innings_played['InningsPlayed']

# total balls faced by each batting pair
balls_faced = ipl.groupby(['batter', 'non-striker']).size().reset_index(name='BallsFaced')

# total runs scored per 100 balls for each batting pair
batting_pairs['StrikeRate'] = (batting_pairs['Runs'] / balls_faced['BallsFaced']) * 100

# Rename columns
batting_pairs.columns = ['Batsman1', 'Batsman2', 'Runs', 'Avg', 'StrikeRate']
````
