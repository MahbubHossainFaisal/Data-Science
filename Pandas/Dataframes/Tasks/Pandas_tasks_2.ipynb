###`Q-4:` Do these below operations:
- Find out the rank based on the "Team" column and save the result by adding a new column named "Rank".
- Change the datatype of this column to integer by using `np.int16`
- Set the index of the DataFrame by using this "Rank" column permanently.
- After that, sort the dataframe based on the "Rank" index.

````python
# code here
df['Rank'] = df['Team'].map(df['Team'].value_counts().rank(method='dense', ascending=True))
df['Rank'] = df['Rank'].astype(int)
df['Rank']
df
````

The values in the 'Rank' column may be coming as NaN (Not a Number) due to the application of the value_counts() function on the 'Team' column. The value_counts() function returns a Series that counts the occurrences of each unique value in the 'Team' column.

If a team appears only once in the 'Team' column, the corresponding count will be 1, and when applying the rank() function with method='dense', it will assign a rank of 1 to that team. However, if a team appears multiple times, the value_counts() function will provide the count for each team, but the rank() function will not assign a rank to any repeated team. This can result in NaN values in the 'Rank' column for teams that have a count greater than 1.

used the **map()** function to assign ranks to teams based on their counts. This approach ensures that teams with the same count are assigned the same rank. Finally, I printed the 'Rank' column of the DataFrame, which now contains the desired ranks for each team.

````python
# code here
ipl_matches = "https://docs.google.com/spreadsheets/d/e/2PACX-1vRy2DUdUbaKx_Co9F0FSnIlyS-8kp4aKv_I0-qzNeghiZHAI_hw94gKG22XTxNJHMFnFVKsO4xWOdIs/pub?gid=1655759976&single=true&output=csv"

ipl = pd.read_csv(ipl_matches)

ipl.head()
````

````python
# changing old names to new names

old_team_names = ['Delhi Daredevils','Kings XI Punjab','Rising Pune Supergiants']
condition = (ipl['Team1'].isin(old_team_names)) |(ipl['Team2'].isin(old_team_names) | (ipl['WinningTeam'].isin(old_team_names)))


mapping = {
    'Delhi Daredevils' : 'Delhi Capitals',
    'Kings XI Punjab' : 'Punjab Kings',
    'Rising Pune Supergiants' : 'Rising Pune Supergiant'
}

ipl.loc[condition,['Team1','Team2','WinningTeam']] = ipl.loc[condition,['Team1','Team2','WinningTeam']].replace(mapping)

ipl[condition]
````

````python
# getting team names

team1 = ipl['Team1']
team2 = ipl['Team2']
teams = pd.concat([team1,team2])

teams_names = teams.unique()
teams_names
````

````python
# number of matches played by teams
teams[teams == 'Mumbai Indians'].value_counts()
````

````python
# Win % of MI
# Only MI matches
mi_matches = ipl[(ipl['Team1'] == 'Mumbai Indians') | (ipl['Team2'] == 'Mumbai Indians')]

# Matches won by MI
mi_win_matches = mi_matches[mi_matches['WinningTeam'] == 'Mumbai Indians']

# win % of MI
win_percentage = (len(mi_win_matches) / len(mi_matches)) * 100
win_percentage

````

````python
# Home win % means Team1 Win%

mi_home_matches = ipl[(ipl['Team1'] == 'Mumbai Indians') ]

# Matches won by MI
mi_win_matches = mi_home_matches[mi_home_matches['WinningTeam'] == 'Mumbai Indians']

# win % of MI
win_percentage = (len(mi_win_matches) / len(mi_home_matches)) * 100
win_percentage
````

````python
# Away win % means Team2 Win%

mi_away_matches = ipl[(ipl['Team2'] == 'Mumbai Indians') ]

# Matches won by MI
mi_win_matches = mi_away_matches[mi_away_matches['WinningTeam'] == 'Mumbai Indians']

# win % of MI
win_percentage = (len(mi_win_matches) / len(mi_away_matches)) * 100
win_percentage
````

````python
# Generalised solution for all teams.
dataframes_array=[]
for i in teams_names:
  
  team_name = i
  matches_played = teams[teams == i].value_counts()
  # win percentage
  matches = ipl[(ipl['Team1'] == i) | (ipl['Team2'] == i)]
  win_matches = matches[matches['WinningTeam'] == i]

  win_percentage = (len(win_matches) / len(matches)) * 100
  
  # Home win percentage
  home_matches = ipl[(ipl['Team1'] == i) ]
  home_win_matches = home_matches[home_matches['WinningTeam'] == i]
  home_win_percentage = (len(home_win_matches) / len(home_matches)) * 100


  # Aways win percentage
  away_matches = ipl[(ipl['Team2'] == i) ]
  away_win_matches = away_matches[away_matches['WinningTeam'] == i]
  away_win_percentage = (len(away_win_matches) / len(away_matches)) * 100

  team_summary = {
      'team': team_name,
      'Match_played': matches_played,
      'Win_percentage': win_percentage,
      'Home_win_percentage': home_win_percentage,
      'Away_win_percentage': away_win_percentage

  }


  df = pd.DataFrame(team_summary)
  dataframes_array.append(df)





````

###`Q-9:` IPL Point Table

Make a function `point_table` which take `season` as parameter and show points table in non-ascendng order of points and in ascending order of team name.

For winning - 2 Ponits; 
For loosing - 0 Point
For not result both team gets 1 points.

Make dataframe which will have
`TeamName` `MatchesPlayed` `MatchesWon` `NoResult` `Points`
make `TeamName` as index.

```
season parametr should be one of these->
['2022', '2021', '2020/21', '2019', '2018', '2017', '2016', '2015', '2014', '2013', '2012', '2011', '2009/10', '2009', '2007/08']
```

Output of two Top 2 in season 2022

```
TeamName    MatchesPlayed	MatchesWon	NoResult	Points
```

````python
# code here
season_2008 = ipl[ipl['Season'] == '2018']

team1 = season_2008['Team1']
team2 = season_2008['Team2']

all_teams = pd.concat([team1,team2])
actual_teams = all_teams.unique()
actual_teams

season_2008['WinningTeam'].value_counts()
````

````python

dashboard = []
for i in actual_teams:
  teamName = i
  matchesPlayed = all_teams[all_teams==i].value_counts().values[0]
  matchesWon = season_2008['WinningTeam'].value_counts()[i] if i in season_2008['WinningTeam'].value_counts() else 0
  matchesLost = matchesPlayed - matchesWon
  points = matchesWon*2

  dic = {
      'team': teamName,
      'MatchesPlayed': matchesPlayed,
      'MatchesWon': matchesWon,
      'MatchesLost': matchesLost,
      'Points': points
  }

  dashboard.append(dic)




df = pd.DataFrame(dashboard)

df
````
