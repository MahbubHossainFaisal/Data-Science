### concat dataframes horizontally

```python
# We have to concate the data of november and december dataframe.

pd.concat([november,december],ignore_index=True)
```

The **ignore_index** parameter is a boolean flag that indicates whether to ignore the original index labels of the concatenated objects and instead create a new sequential index. Here's how it works:

When ignore_index is set to False (the default), the concatenated DataFrame will retain the original index labels from each individual DataFrame. This means that the resulting DataFrame will have **non-unique index labels** if the original DataFrames had overlapping or duplicate index values.

When ignore_index is set to True, the concatenated DataFrame will ignore the original index labels and instead create a new sequential index from 0 to N-1, where N is the total number of rows in the concatenated DataFrame. This can be useful when the original index labels are not meaningful or when you want a fresh index for the combined data.

## Merge (Like Database JOINS)

```python
# Inner Join

registered = pd.concat([november,december],ignore_index=True)
```

```python
students
```

```python
students.merge(registered,how='inner',on='student_id')
```

```python
# left join

courses
```

```python
courses.merge(registered,how='left',on='course_id')
```

```python
temp_df = pd.DataFrame({
    'student_id':[26,27,28],
    'name':['Faisal','Asif','Ratul'],
    'partner':[28,26,17]
})

students = pd.concat([students,temp_df],ignore_index=True)
```

```python
# right join
students
```

```python
students.merge(registered,how='right',on='student_id')
```

```python
# outer join

students.merge(registered,how='outer',on='student_id').tail(10)
```

```python
# 1. find total revenue generated

registered.merge(courses,how='inner',on='course_id')['price'].sum()
```

```python
# 2. find month by month revenue

monthly_registration = pd.concat([november,december],keys=(['nov','dec'])).reset_index()
monthly_registration
```

```python
monthly_registration.merge(courses,how='inner',on='course_id').groupby('level_0')['price'].sum()
```

```python
# 3. Print the registration table
# cols -> name -> course -> price

registered.merge(students,how='inner',on='student_id').merge(courses,how='inner',on='course_id')[['name','course_name','price']]
```

```python
# Plot bar chart for revenue/course (Which course made how much)

course_revenues = registered.merge(courses,how='inner',on='course_id').groupby('course_name')['price'].sum()
course_revenues
```

```python
course_revenues.plot(kind='bar',ylabel='revenue')
```

```python
# 5. find students who enrolled in both the months

common_student_id = np.intersect1d(november['student_id'],december['student_id'])
common_student_id
```

```python
students[students['student_id'].isin(common_student_id)]
```

```python
# 6. find course that got no enrollment
# courses['course_id']
# regs['course_id']

course_enrollment= courses.merge(registered,how='left',on='course_id')

course_enrollment
```

```python
course_enrollment[course_enrollment['student_id'].isna()]['course_name']
```

```python
# 7. find students who did not enroll into any courses



student_id_list = np.setdiff1d(students['student_id'],registered['student_id'])
students[students['student_id'].isin(student_id_list)][['student_id','name']]
```

```python
# student who didnot enroll
# another way
students_enrollment = students.merge(registered,how='left',on='student_id')
students_enrollment[students_enrollment['course_id'].isna()][['student_id','name']]
```

```python
# 8. Print student name -> partner name for all enrolled students
# self join
students.merge(students,how='inner',left_on='partner',right_on='student_id')[['name_x','name_y']]
```

```python
# 9. find top 3 students who did most number enrollments
registered.merge(students,on='student_id').groupby(['student_id','name'])['name'].count().sort_values(ascending=False).head(3)
```

```python
# 10. find top 3 students who spent most amount of money on courses
registered.merge(students,on='student_id').merge(courses,on='course_id').groupby(['student_id','name'])['price'].sum().sort_values(ascending=False).head(3)
```

```python
# Alternate syntax for merge
# students.merge(regs)

pd.merge(students,registered,how='inner',on='student_id')
```
