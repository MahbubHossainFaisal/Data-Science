###`Q-8:` Team1 vs Team2 Dashbord
Create a function which will take two string(name of two teams) as input. Show win Loss record between them and player getting most player of the match award in matches between these two teams.
```
team1_vs_team2('Kolkata Knight Riders','Chennai Super Kings')
```

````python
# code here

def team1_vs_team2(t1,t2):
  # total number of team t1 matches
  t1_matches  = ipl[(ipl['Team1'] == t1) | (ipl['Team2'] == t1)]
  # total number of team t2 matches
  t2_matches  = ipl[(ipl['Team1'] == t2) | (ipl['Team2'] == t2)]

  # total lost matches for t1 
  t1_lost_matches = t1_matches[t1_matches['WinningTeam'] != t1]

  # total lost matches for t2
  t2_lost_matches = t2_matches[t2_matches['WinningTeam'] != t2]

  # t1 win matches
  t1_win_matches = len(t1_matches) - len(t1_lost_matches)

 

  # t2 win matches
  t2_win_matches = len(t2_matches) - len(t2_lost_matches)

  print(f'Matches won by {t1}',t1_win_matches)
  print(f'Matches lost by {t1}',len(t1_lost_matches))
  print(f'Matches won by {t2}',t2_win_matches)
  print(f'Matches lost by {t2}',len(t2_lost_matches))


  # Player getting most award in t1 team
  player_of_the_match_count_t1 = t1_matches['Player_of_Match'].value_counts()
  most_awards_t1 = player_of_the_match_count_t1[player_of_the_match_count_t1 == player_of_the_match_count_t1.max()]

  print(f'Most award winning player from {t1} \n', most_awards_t1.to_string(header=False))

  player_of_the_match_count_t2 = t2_matches['Player_of_Match'].value_counts()
  most_awards_t2 = player_of_the_match_count_t2[player_of_the_match_count_t2 == player_of_the_match_count_t2.max()]

  print(f'Most award winning player from {t2}\n', most_awards_t2.to_string(header=False))




team1_vs_team2('Mumbai Indians','Chennai Super Kings')
````
