# Colored Scatterplots

```python
temp = pd.read_csv('D:\\Study\\Datasets\\Matplotlib_data\\datasets-session-24\\iris.csv')

iris = temp.copy()
```

```python
iris['Species'].unique()
```

We want to make a scatterplot which represents the sepal length dn petal length of different species. And we want the scatters
to be in different color for different species.

But as the species are in text or string format we have to replace them with numbers in order to show their sepal length and
petal length in different color

```python
iris['Species'] = iris['Species'].replace({'Iris-virginica':1,'Iris-versicolor':2,'Iris-setosa':3})
iris['Species']
```

Now species values has converted into numbers

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'])
```

This is how we can change the colors for different species. Now if we don't like the default colors we can add more themes
using __cmap__ parameter

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'], cmap='jet')
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'], cmap='winter')
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'], cmap='summer')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

You can change the opacity by using __alpha__ parameter

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'], cmap='summer',alpha=0.6)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

If we want to __resize__ the graph plot, we have to use __plt.figure(figsize=(x,y))__

```python
plt.figure(figsize = (10,7))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'], cmap='summer',alpha=0.6)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

## Annotation

```python
temp = pd.read_csv('D:\\Study\\Datasets\\Matplotlib_data\\datasets-session-24\\batter.csv')

batter = temp.copy()
```

```python
batter.head()
```

Now we will take 15 random batsman data from the dataframe

```python
sample_df = batter.head(100).sample(25,random_state=5)
sample_df
```

```python
plt.scatter(sample_df['avg'],sample_df['strike_rate'])
```

Now it is not understandable which bubble represent which player in the chart. 

So now we can label them using __plt.text()__ which is called annotation

```python
sample_df.shape[0] # this the row number
```

```python
plt.figure(figsize=(10,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'])

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
    
    
```

Now I also want to visualize the runs using the dots. The more runs any batsman have , the dots will be bigger.

for that we will use __s__ parameter in __plt.scatter()__

```python
plt.figure(figsize=(20,15))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
    
    
```

## Horizontal and Vertical Lines

We can do horizontal and vertical lines to seperate different range.

```python
plt.figure(figsize=(20,15))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

plt.axhline(130,color='red')
plt.axvline(30,color='red')

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

Here in the upper right corner we can see the players who has avg runs above 30 and strike rate above 130.So in a glimpse we can find the best players.

Now lets make a 2 graphs in a single figure

```python
fig, ax = plt.subplots(nrows=2,ncols=1)
```

```python
fig
```

Fig is representing the structure

```python
ax
```

ax is representing the two chart areas as array. where ax[0] will indicate the first chart area, ax[1] will indicate the 2nd chart area

```python
fig, ax = plt.subplots(nrows=2,ncols=1)
ax[0].scatter(batter['avg'],batter['strike_rate'])
ax[1].scatter(batter['avg'],batter['runs'])

ax[0].set_title('Avg VS Strike Rate')
ax[1].set_title('Avg VS Runs')

ax[0].set_ylabel('Strike Rate')
ax[1].set_ylabel('Runs')
```

Now we can see that, the middle values got mixed with the title of the second graph. What we can do is, as both the graph has
same axis value (avg), we use __sharex__ parameter oin the subplot function. Then the axis values will be shown only the bottom graph

```python
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True)
ax[0].scatter(batter['avg'],batter['strike_rate'])
ax[1].scatter(batter['avg'],batter['runs'])

ax[0].set_title('Avg VS Strike Rate')
ax[1].set_title('Avg VS Runs')

ax[0].set_ylabel('Strike Rate')
ax[1].set_ylabel('Runs')
```

### 4 graphs

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].scatter(batter['avg'],batter['strike_rate'])
ax[0,1].scatter(batter['avg'],batter['runs'])
ax[1,0].hist(batter['avg'])
ax[1,1].hist(batter['runs'])
```

### 3 graphs (Another way to plot subplots)

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batter['avg'],batter['strike_rate'],color='red')

ax2 = fig.add_subplot(2,2,2)
ax2.hist(batter['runs'])

ax3 = fig.add_subplot(2,2,3)
ax3.hist(batter['avg'])
```

# 3D Surface Plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.log(xx)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
