### If we have outliers in data, graph can become like this

```python
# plotting a simple function
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
```

Now suppose , you have a range of your data in mind and you don't want to make outliers hamper your chart. So, you can limit the range of your x axis or y axis.You can use the below method of lim()

```python
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.ylim(0,75000)
```

### if you want grid in your chart you can use grid()

```python
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.ylim(0,75000)
plt.grid()
```

In anaconda, the graphs can be seen without using __plt.show()__. But in other platforms we have to use plt.show() to actually visualize the graph

### plt.plot vs plt.scatter

__plt.plot()__: This method is primarily used to create line plots but can also be used to create scatter plots by specifying marker styles. It connects the data points with lines by default, making it suitable for visualizing continuous data or data with a sequential relationship.

__plt.scatter()__: This method is specifically designed for creating scatter plots. It displays individual data points as markers without connecting them with lines. It provides additional flexibility, such as customizing marker styles, sizes, colors, and assigning different colors to different groups of data points. Scatter plots are useful for visualizing relationships between two variables, identifying clusters or patterns, and detecting outliers.

In summary, plt.plot() is primarily for line plots but can be used for scatter plots with marker styles, while plt.scatter() is specifically designed for scatter plots and offers more flexibility and functionality for customizing markers and exploring relationships between variables.

Also for faster response plt.scatter() is more efficient in terms of large dataset

If we plot three bar charts together , it will come one above another. So all three can't be seen.
So we have to come up with some shifting technique so that all the bars can be visible.

Also we convert the x axis values to numbers because we can't apply shifting without converting the x axis values to numbers. because we are adding another number or substructing another number in order to shift

```python
# first bar left shift by .2 , second bar no shift, third bar shift right side .2
plt.bar(np.arange(batsman.shape[0])-0.2,batsman['2015'],width=.2)
plt.bar(np.arange(batsman.shape[0]),batsman['2016'],width=.2)
plt.bar(np.arange(batsman.shape[0])+0.2,batsman['2016'],width=.2)
```

But here we can see that, we cannot see the batsmen names in the x axis. So we can use another method __xticks()__ to see that

```python
# first bar left shift by .2 , second bar no shift, third bar shift right side .2
plt.bar(np.arange(batsman.shape[0])-0.2,batsman['2015'],width=.2)
plt.bar(np.arange(batsman.shape[0]),batsman['2016'],width=.2)
plt.bar(np.arange(batsman.shape[0])+0.2,batsman['2016'],width=.2)

plt.xticks(np.arange(batsman.shape[0]),batsman['batsman'])
```

There are many techincal info coming. To remove those use plt.show()

```python
# first bar left shift by .2 , second bar no shift, third bar shift right side .2
plt.bar(np.arange(batsman.shape[0])-0.2,batsman['2015'],width=.2)
plt.bar(np.arange(batsman.shape[0]),batsman['2016'],width=.2)
plt.bar(np.arange(batsman.shape[0])+0.2,batsman['2016'],width=.2)

plt.xticks(np.arange(batsman.shape[0]),batsman['batsman'])
plt.show()
```

__Changing Colors__:

```python
plt.bar(np.arange(batsman.shape[0])-0.2,batsman['2015'],width=.2, color = 'black')
plt.bar(np.arange(batsman.shape[0]),batsman['2016'],width=.2, color='orange')
plt.bar(np.arange(batsman.shape[0])+0.2,batsman['2016'],width=.2, color = 'red')

plt.xticks(np.arange(batsman.shape[0]),batsman['batsman'])
plt.show()
```

### Logarithm scale

```python
plt.hist(arr,bins=[10,20,30,40,50,60,70],log=True)
```

The __log__ parameter in the plt.hist() function in Matplotlib is used to set the histogram axis to a log scale. This means that the y-axis will be scaled logarithmically, which can be useful for visualizing data that has a wide range of values.

For example, let's say you have a dataset of population sizes for different countries. The population sizes of these countries can vary widely, from a few thousand people to hundreds of millions of people. If you plot a histogram of this data without using the log parameter, the y-axis will be very compressed, and it will be difficult to see the differences in the population sizes.

# Pie Chart

- Univariate/Bivariate Analysis

- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
marks = [85,33,91,95,65]
subjects = ['English','Maths','Science','ICT','Social Science']

plt.pie(marks,labels=subjects)
plt.show()
```
