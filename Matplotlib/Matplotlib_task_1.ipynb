### **`Problem-1:`** Draw a line plot of which, the x-axis is the "Year" and the y-axis is sum of "PM2.5" of two countries Iran and China. 

```python
# code here
df = pd.read_csv('https://tinyurl.com/2fe6vz4u')
df.head()
```

```python
Iran_data = df[df['Country'] == 'Iran'].groupby('Year')['PM2.5'].sum()

Chinese_data = df[df['Country'] == 'China'].groupby('Year')['PM2.5'].sum()
```

```python
Iran_data
```

```python
plt.plot(Iran_data.index,Iran_data.values,label='Iran')
plt.plot(Chinese_data.index,Chinese_data.values,label='China')

plt.title('PM2.5 Over the years (India VS China)')
plt.xlabel('Year')
plt.ylabel('PM2.5')
plt.xticks(df['Year'].value_counts().index) # xticks change the values of x axis.
plt.show()
```

### **`Problem-2:`** Draw a histogram of the  column "PM10" of which the y-axis represents the probability (see the documentation how to draw the probability). 

A histogram is a graphical representation of the distribution of a dataset. It provides a visual representation of the frequencies or counts of values falling into different intervals, known as bins.

The x-axis of a histogram represents the range of values in the dataset, divided into intervals or bins. The y-axis represents the frequency or count of values that fall within each bin.

Histograms are useful for understanding the shape, central tendency, and spread of a dataset. They can reveal information about the distribution of the data, such as whether it is symmetric, skewed, or bimodal. Histograms can also help identify outliers and understand the frequency or density of values in different ranges.

By examining the shape of a histogram, you can gain insights into the underlying patterns and characteristics of the data. Common patterns in histograms include normal distributions, skewed distributions (positively or negatively), uniform distributions, and multimodal distributions.

Histograms are widely used in data analysis, statistics, and data visualization as they provide a visual summary of the distribution of a dataset. They are particularly useful when working with continuous or numerical data.

```python
plt.hist(df['PM10'],bins=list(range(1,400,20)),color='black')
plt.xlabel('Bins')
plt.ylabel('Probability')
plt.title('Histogram of PM10')
plt.grid()
plt.show()
```

### `Problem-10` Plot Stacked Bar chart quarter wise for each product.

```python
final_df
```

```python
# Create a new 'quarter' column based on the 'month_number'
df['quarter'] = pd.to_datetime(df['month_number'], format='%m').dt.quarter

df.head()
```

```python
# Group the data by quarter and sum the sales for each product
quarterly_sales = df.groupby('quarter')[['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']].sum()
quarterly_sales
```

```python

# Plot the stacked bar chart
quarterly_sales.plot(kind='bar', stacked=True)

# Add labels and title
plt.xlabel('Quarter')
plt.ylabel('Sales')
plt.title('Stacked Bar Chart - Quarterly Sales by Product')
plt.xticks(rotation=0)
# Show the chart
plt.show()
```
