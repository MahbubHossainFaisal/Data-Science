## Data Interpolation Technique

Data interpolation is a mathematical technique used to estimate values within a set of known data points. It involves fitting a curve or function to the existing data in such a way that it can predict or estimate values at intermediate points. Interpolation is commonly used in various fields, including computer graphics, geophysics, and finance, to fill in missing or unobserved data. The choice of interpolation method, such as linear, polynomial, or spline interpolation, depends on the nature of the data and the desired level of accuracy. However, it's essential to be cautious when extrapolating data beyond the observed range, as it can lead to inaccurate predictions.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Linear Interpolation

```python
# Taking two data points

x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

plt.scatter(x,y)
```

### Interpolating the data using linear interpolation

```python
# creating new x values
new_x = np.linspace(1,5,10)
y_interp = np.interp(new_x,x,y) # liner interpolate
y_interp
```

you are performing linear interpolation between two sets of data points to estimate values for new_x based on the relationship between x and y.

1. You have two sets of data points:
   - x = [1, 2, 3, 4, 5]
   - y = [2, 4, 6, 8, 10]

2. You create a new set of x values, `new_x`, using `np.linspace`:
   - `new_x = np.linspace(1, 5, 10)` generates an array of 10 evenly spaced values between 1 and 5.

3. The `np.interp` function is then used to perform linear interpolation. It takes `new_x` as the first argument, the original x values `x` as the second argument, and the original y values `y` as the third argument.

4. For each value in `new_x`, the `np.interp` function calculates a corresponding y value (`y_interp`) by estimating it based on the linear relationship between the closest data points in `x` and `y`. It uses a straight line to connect these two closest data points and determines the y value at the position of `new_x`.

5. The resulting `y_interp` array contains the estimated y values for each corresponding value in `new_x` based on linear interpolation.

So, `y_interp` will contain the interpolated values for `new_x` based on the linear relationship between the original data points in `x` and `y`. This allows you to estimate y values at points that were not originally in your dataset.

```python
plt.scatter(new_x,y_interp,color='r')
plt.scatter(x,y)
```

The red points are newly added through linear interpolation. We can see that more values have been generated between the line points (see previous figure ).That's how linear interpolation upsample data

## Cubic Interpolation
We will use Scipy for this

```python
from scipy.interpolate import interp1d
```

```python
# Taking two data points

x = np.array([1,2,3,4,5])
y = np.array([1,8,27,64,125])

plt.scatter(x,y,color='r')
```

```python
cubic_interp_func = interp1d(x,y,kind='cubic',)
```

```python
# creating new x values
new_x = np.linspace(1,5,10)
y_interp = cubic_interp_func(new_x) # liner interpolate
y_interp
```

```python
plt.scatter(new_x,y_interp)
plt.scatter(x,y,color='r')
```

The blue points are newly added through cubic interpolation.

## Polynomial Interpolation

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 4, 9, 16, 25])
```

```python
plt.scatter(x,y)
```

```python
# interpolate the data using polynomial interpolation
p=np.polyfit(x,y,3)
```

```python
p
```

```python
x_new = np.linspace(1, 5, 10) # create new x values
y_interp = np.polyval(p, x_new) # interpolate y values
```

```python
plt.scatter(x_new,y_interp)
plt.scatter(x,y)
```

Here the orange points are newly added through polynomial interpolation.
